```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error, r2_score

original_dataset = pd.read_csv('my_original_data.csv')

target_variable = 'median_house_value'

# Function to perform kNN Regression and plot MSE vs k curve
def knn_regression_and_plot_mse(X, y, dataset_name):
    
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

    # Initialize variables
    mse_values = []
    r2_values = []

    k_values = range(1, 31)

    # Iterate over different k values
    for k in k_values:
        knn_regressor = KNeighborsRegressor(n_neighbors=k)

        knn_regressor.fit(X_train, y_train)

        y_pred = knn_regressor.predict(X_test)

        # i calculated MSE and R-squared values
        mse = mean_squared_error(y_test, y_pred)
        r2 = r2_score(y_test, y_pred)

        mse_values.append(mse)
        r2_values.append(r2)

    optimal_k = k_values[np.argmin(mse_values)]

    plt.figure(figsize=(10, 6))
    plt.plot(k_values, mse_values, marker='o')
    plt.title(f'MSE vs k for {dataset_name} Dataset')
    plt.xlabel('Number of Neighbors (k)')
    plt.ylabel('Mean Squared Error (MSE)')
    plt.grid(True)
    plt.show()

    print(f"Optimal k for {dataset_name} dataset: {optimal_k}")
    print(f"MSE for optimal k: {mse_values[optimal_k - 1]}")
    print(f"R-squared value for optimal k: {r2_values[optimal_k - 1]}")


X_original = housing_df[['longitude', 'latitude', 'housing_median_age', 'total_rooms',
       'total_bedrooms', 'population', 'households', 'median_income',
       'median_house_value']]

X_outlier_removed = filtered_dataset_total_rooms[['longitude', 'latitude', 'housing_median_age', 'total_rooms',
       'total_bedrooms', 'population', 'households', 'median_income',
       'median_house_value' ]]
X_mean_imputed = df_copy3[['longitude', 'latitude', 'housing_median_age', 'total_rooms',
       'total_bedrooms', 'population', 'households', 'median_income',
       'median_house_value']]
X_median_imputed = df_copy4[['longitude', 'latitude', 'housing_median_age', 'total_rooms',
       'total_bedrooms', 'population', 'households', 'median_income',
       'median_house_value']]


y_original = housing_df['median_house_value']
y_outlier_removed = filtered_dataset_total_rooms['median_house_value']
y_mean_imputed = df_copy3['median_house_value']
y_median_imputed = df_copy4['median_house_value']




# For the original dataset
knn_regression_and_plot_mse(X_original, y_original, 'Original')

# For the outlier-removed dataset
knn_regression_and_plot_mse(X_outlier_removed, y_outlier_removed, 'Outlier Removed')

# For the mean-imputed dataset
knn_regression_and_plot_mse(X_mean_imputed, y_mean_imputed, 'Mean Imputed')

# For the median-imputed dataset
knn_regression_and_plot_mse(X_median_imputed, y_median_imputed, 'Median Imputed')
```

    * Fit a linear regression model to train data. Predict the test set and calculate the MSE value.

```python
from sklearn.linear_model import LinearRegression

def linear_regression_and_plot(X, y, dataset_name):
    # i splitted the data into train and test sets
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

    linear_regressor = LinearRegression()

    linear_regressor.fit(X_train, y_train)

    y_pred = linear_regressor.predict(X_test)

    # Calculate MSE
    mse = mean_squared_error(y_test, y_pred)

    # Plot the actual vs. predicted values
    plt.figure(figsize=(10, 6))
    plt.scatter(y_test, y_pred, color='blue')
    plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], linestyle='--', color='red', linewidth=2, label='Ideal Prediction')
    plt.title(f'Linear Regression for {dataset_name} Dataset')
    plt.xlabel('Actual Values')
    plt.ylabel('Predicted Values')
    plt.legend()
    plt.show()

    # Print the MSE value
    print(f"MSE for {dataset_name} dataset: {mse}")

# Call the function for each version of the dataset
linear_regression_and_plot(X_original, y_original, 'Original')
linear_regression_and_plot(X_outlier_removed, y_outlier_removed, 'Outlier Removed')
linear_regression_and_plot(X_mean_imputed, y_mean_imputed, 'Mean Imputed')
linear_regression_and_plot(X_median_imputed, y_median_imputed, 'Median Imputed')

```
